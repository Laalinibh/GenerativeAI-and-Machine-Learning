# A task has been created in Asana tool and the same is emailed to the user

```python
import requests
from datetime import datetime

def create_task(api_token, workspace_id, project_id, task_name, assignee_id, due_date):
    headers = {
        'Authorization': f'Bearer {api_token}',
        'Content-Type': 'application/json'
    }

    data = {
        'data': {
            'workspace': workspace_id,
            'projects': [project_id],
            'name': task_name,
            'assignee': assignee_id,
            'due_on': due_date.strftime('%Y-%m-%d')
        }
    }

    response = requests.post('https://app.asana.com/api/1.0/tasks', headers=headers, json=data)

    if response.status_code == 201:
        return response.json()['data']
    else:
        print(f"Error: {response.status_code} - {response.text}")
        return None


# Example usage
api_token = '1/1204695278604361:eab381334a69c228578926646fb70ab1'
workspace_id = '76786340109967'
project_id = '1204695278739271'
task_name = 'Meeting Scheduler'
assignee_id = '1204695278604376'
due_date = datetime(2023,5,30)  # Format: YYYY-MM-DD

task_id = create_task(api_token, workspace_id, project_id, task_name, assignee_id, due_date)
if task_id:
    print(f"Task created with ID: {task_id}")
```

```python
//send emails to panel

import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

def send_emails(emails, subject, message):
    # Email configuration
    smtp_host = 'YOUR_SMTP_HOST'
    smtp_port = 587
    smtp_username = 'YOUR_SMTP_USERNAME'
    smtp_password = 'YOUR_SMTP_PASSWORD'
    sender_email = 'SENDER_EMAIL_ADDRESS'
    
    for recipient_email in emails:
        # Create a multipart message
        msg = MIMEMultipart()
        msg['From'] = sender_email
        msg['To'] = recipient_email
        msg['Subject'] = subject

        # Add the message content
        body = message  # Replace with your template or custom message
        msg.attach(MIMEText(body, 'plain'))

        # Create an SMTP session and send the email
        with smtplib.SMTP(smtp_host, smtp_port) as server:
            server.starttls()
            server.login(smtp_username, smtp_password)
            server.send_message(msg)

        print(f"Email sent to: {recipient_email}")

# Example usage
email_list = ['recipient1@example.com', 'recipient2@example.com', 'recipient3@example.com']
email_subject = 'Meeting Invitation'
email_message = '''
Dear Participant,

You are invited to a meeting on Tuesday at 2:00 PM.

Please confirm your availability.

Best regards,
Your Name
'''

send_emails(email_list, email_subject, email_message)
```
