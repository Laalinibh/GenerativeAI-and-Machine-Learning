```python
import pandas as pd
import spacy
from sklearn.metrics.pairwise import cosine_similarity


# Load spaCy model with word vectors (medium-sized model)
nlp = spacy.load('en_core_web_md')

# Function to get embeddings for a sentence using spaCy
def get_embedding(sentence, model):
    doc = model(sentence)
    return doc.vector
#print(type(Attendee_df['interest'].values))
v =[]
for i in Attendee_df['interest'].values:
    v.append(get_embedding(i, nlp))

#print(v)
Events_df['Event_Name'] = Events_df['Event_Name'].astype(str)
k =[]
for j in Events_df['Event_Name'].values:
    k.append(get_embedding(j, nlp))
    
#print(k)


# Apply the function to the 'interest' column in Attendee_df
Attendee_df['embedding'] = v
Events_df['embedding'] = k
# Apply the function to the 'other_column' in df_B


# Calculate cosine similarity between embeddings
similarity_matrix = cosine_similarity(list(Attendee_df['embedding']), list(Events_df['embedding']))

# Create a DataFrame with similarity scores
similarity_df = pd.DataFrame(similarity_matrix, index=Attendee_df.index, columns=Events_df.index)

print(similarity_df)
```
